# 실습: while 문
* while문을 이용한 무한 반복 실행해보기
* [99, 66, 8, 74, 'ABC, '456']리스트에서 하나씩 값을 출력하는 코드를 만들어보자
* 이 때, while 무한 반복문을 사용해보자
* 에러가 발생할 경우, 원인을 파악한다
* 리스트의 인덱싱을 이용한다
* 리스트 또는 인덱싱에 다른 조치를 취하지 않는다

```python
list = [99, 66, 8, 74, 'ABC', '456']
int = 0
while True:
    print(list[int])
    int += 1
# while문이 리스트 값의 범위를 벗어남
```

```python
list = [99, 66, 8, 74, 'ABC', '456']
while 1:
    print(list.pop())
```

코드의 가독성이 떨어지거나, 코드의 규약이 없는 경우, 코드의  결합도가 높은 경우, 시간이 없어서 땜빵 코드를 작성한 경우, 이런 여러 가지 원인들로 인해서 화석처럼 굳어진 코드를 흔히 '레거시 코드'라고 표현한다.

# 이중루프
* for문, while문은 모두 이중루프 이상이 가능하다.

```python
list_ex = []

for i in range(10):
    for j in range(10):
        list_ex.append([1, 2, 3])
```

```python
list_ex
```

```python
list_ex2 = []

for i in range(100):
    list_ex2.append([1, 2, 3])
```

```python
list_ex == list_ex2 #논리적으로, 결과적으로 봤을 때 단일 루프임. 삼중루프를 이중이나 단일로 바꿀 수 있다면 변환해야함.
#논리적으로 변경할 수 있다면 루프를 최소한으로 써야 됨.
```

```python
list_ex3 = []

for i in range(10):
    list_ex3.append([1])
    
    for j in range(10):
        list_ex3.append([2, 3])
```

# 실습
* 이중 루프 만들기
* 루프문 5개를 이용한 이중 루프를 만들자
* 단, 루프문 3개는 서로 관련이 있어야 한다 (같은 변수를 공유하거나, 해당 변수의 값을 받아 쓴다거나)
* 어떤 루프문 2개는 서로 관련이 없어야 한다 (변수 등을 공유하지 않고, 서로의 값이 영향을 끼치지 못함)
* 하나 이상의 루프문은 다른 루프문 하나 이상을 감싼 형태여야 한다
* 두 덩어리의 이중루프가 있어도 무방하다
* 예제의 종류 및 연산방법은 수강생이 정한다
    

```python
for a in range(5):
    for b in range (5, a, -1):
        print(' ', end="")
    print(end="")
    
    for c in range (0, a, 1):
        print('*', end="")
    print(end="\n")
    


for b in range(5):
    for d in range (5, b, -1):
        print('*', end="")
    print(end="\n")
```

```python
for i in range(1, 11):
    for j in range (1, 10):
        print(i, '* ',  end = "")
        print(j, '= ', end = "")
        print(i*j, end = '\n')
```

```python
for a in range(3):
    for b in range (3, a, -1):
        print(' ', end="")
    
    for c in range (0, a, 1):
        print('*', end="")
    
    for c in range (0, a, 1):
        print('*', end="")
    
    for b in range (3, a, -1):
        print(' ', end="")
        
    for b in range (3, a, -1):
        print(' ', end="")
    
    for c in range (0, a, 1):
        print('*', end="")
    
    for c in range (0, a, 1):
        print('*', end="")
    
    for b in range (3, a, -1):
        print(' ', end="")
    
    
    print(end="\n")
    


for b in range(6):
    for e in range (0, b, 1):
        print(' ', end="")
        
    for d in range (6, b, -1):
        print('*', end="")   
    
    for d in range (6, b, -1):
        print('*', end="")
    print(end="\n")
```

```python
# 위 코드를 참조해, 사용자가 지정한 시간만큼 일시정지하도록 처리하기
# 단, 일시정지하는 동안 매 0.5초마다 '일시정지'라고 출력한다.

# 5번째 루프에서 사용자가 지정한 시간만큰 일시정지하도록 처리하기
x = int(input('초를 입력하세요: '))

for i in range(10):
    if i == 5:
        for j in range(x*2):
            time.sleep(0.5)
            print('일시정지')
    else: 
        print('*'*i)
    
  
```

# 실습: 데이터프레임 인덱싱 후 값 연산하기
* df 데이터프레임의 값을 이용한다
* 현실 숫자로 4, 10000 번째 값과 2, 55023번째 값을 더해보자
* 계산 결과는 res 변수에 저장한다

---
* 해당 연산을 이용해서 특정 범위를 숫자로 지정하면 모든 값을 더하도록 프로그램을 만든다
* 단, index 범위는 방향으로만 지정한다 (column 범위는 자유롭게 설정)

```python
pd.options.display.float_format = '{:.5f}'.format # 지수 변환을 실수로 보여주는 코드, 소수점 5자리까지 보여줌
res1 = 0
res2 = 0
for i in range(4-1, 10000-1):
    res1 += df.iloc[i]
    
for j in range(2-1, 55023-1):
    res2 += df.iloc[j]
    
print(res1 + res2)
```

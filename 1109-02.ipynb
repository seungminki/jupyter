```python
# 효율적인 코드
def sum_all(x, y, z, w):
    sum_1 = x + y
    sum_2 = sum_1 + z
    sum_4 = sum_2 * 2 + sum_1
    sum_5 = sum_4 * w
        
    return sum_5 

def list_return(x, z, w):
    return [i+x+z for i in range(w)]

x = int(input('number 1: '))
y = int(input('number 2: '))
z = int(input('number 3: '))
w = int(input('number 4: '))

num1_res = sum_all(x, y, z, w)
num3_res = list_return(num1_res, z, w)
print(num3_res)
```

# lambda 함수
* lambda x: 형식으로 이용
* 불특정 대상 x를 의미
* df['WT1'].apply(lambda x: x+1) 식으로 이용할 경우 pandas serial(= series) 로 변환됨
* (lambda x: x+1)(456) 식으로 사용할 경우 int로 반환됨
* 위 예시를 제외한 방법으로 list-like object를 만들어보자

# 실습
* 위 정보들을 이용해 A, C, G, T가 몇개 들어있는지 알 수 있는 데이터 프레임을 만들자

```python
tnfa_a = 0
tnfa_c = 0
tnfa_g = 0
tnfa_t = 0


for i in range (936):
    if tnfa_list[i] == 'A':
        tnfa_a += 1
        
    elif tnfa_list[i] == 'C':
        tnfa_c += 1
        
    elif tnfa_list[i] == 'G':
        tnfa_g += 1
        
    else:
        tnfa_t += 1
    
print("A: ", tnfa_a)
print("C: ", tnfa_c)
print("G: ", tnfa_g)
print("T: ", tnfa_t)
```

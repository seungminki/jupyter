# 실습 - 주사위 게임 만들기
* numpy 패키지를 이용하여 주사위 게임 만들기
1. 컴퓨터는 두 숫자를 랜덤하게 받아낸다.
2. 두 숫자는 육면체 주사위 눈의 그것과 같다
3. 처음에는 사용자에게 값을 알려주지 않고, 첫번째 주사위가 두 번째 주사위의 눈보다 큰지 물어본다
4. 사용자에게 큰지 작은지 응답을 받고, 맞췄는지 출력해준다
5. 해당 프로세스를 반복해서 할 수 있도록 한다 (입력에 따라 종료할 수도 있게 한다)

---
# 혹시나 다 하고 시간이 남는 분들은 아래 프로그램으로 확장해 봅시다
* 주사위 면을 12면체로 확장합니다
* 굴릴 주사의 개수를 사용자가 선택하도록 합니다
* 컴퓨터는 랜덤한 두 주사위를 선택하고, 그 주사위 두개의 대소를 묻습니다

```python
# 기승민 실습 파일
while True:
    
    x = int(input('주사위 게임을 하고싶다면 1을, 종료하고 싶다면 0을 눌러주세요: '))
    random_num = np.random.randint([5, 5])
    print(random_num)

    if x == 1:
        
        if random_num[0] > random_num[1]:
            print('첫번째 주사위의 숫자가 더 큽니다.')
        elif random_num[0] < random_num[1]:
            print('두번째 주사위의 숫자가 더 큽니다.')
        else:
            print('두 주사위의 숫자가 같습니다.')
        
        
        
        a, b = map(int,input('숫자 두개를 띄어쓰기로 입력해주세요(1~6까지)').split())
        
            
        if random_num[0] == a-1:
            
            if random_num[1] == b-1:
                print('맞췄습니다!')
                
            else:
                print('틀렸습니다')
                    
        else:
            print('틀렸습니다')
            
        
    elif x == 0:
        break
        
    else:
        print('다시 입력해주세요')
```

```python
# 손충모님 실습 파일
while True:
    
    x = int(input('주사위 게임을 하고싶다면 1을, 종료하고 싶다면 0을 눌러주세요: '))
    
    if x == 1:
        
        rd = np.random.randint(1, high= 13, size=2)
        iv = input("Is the first number bigger than the second number?(Y/N): ")

        if (iv=='Y' and list(rd)[0]>list(rd)[1]) or (iv=='N' and list(rd)[0]<list(rd)[1]):
            print("correct")

        else:
            print("wrong")
            print("The first number is",rd[0],"The second number is",rd[1])
            
    elif x == 0:
        break
        
    else:
        print('다시 입력해주세요')
```

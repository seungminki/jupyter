# h)
*  남자와 여자 승객 사이에서 티켓 가격(Fare)가 서로 통계적으로 차이를 보이는 지 알아보자(t-test 이용)

```python
test_data8 = test_data7.copy()
```

```python
import numpy as np
from scipy import stats
```

```python
test_male = []
test_female = []

for i in range(418):
    if (test_data8['Sex'][i] == 'female'):
        test_female += [test_data8['Fare'][i]]
    elif (test_data8['Sex'][i] == 'male'):
        test_male += [test_data8['Fare'][i]]
        
print(test_male)
print(test_female)

test_male2 = test_male[:9]
test_female2 = test_female[:9]

print(test_male2)
print(test_female2)
```

```python
# sample size
N = 10.0

#Data of group 1, 2

a = np.array(test_male2)
b = np.array(test_female2)

var_a = a.var(ddof=1.0)
var_b = b.var(ddof=1.0)

#std deviation
s = np.sqrt((var_a + var_b)/2.0)

## Calculate the t-statistics
t = (a.mean() - b.mean())/(s*np.sqrt(2.0/N))

## Compare with the critical t-value
#Degrees of freedom
df = len(a) + len(b) - 2.0

#p-value after comparison with the t
p = stats.t.cdf(t,df=df)

print("t = " + str(t))
print("p = " + str(2.0*p))
```

```python
def ten_percent(x):
    if x > (project_df['Fare'].max() - project_df['Fare'].min()) / 10 * 9:
        print('10%')
        return '10%'
    else:
        return x
```

```python
def strlen(x, y, z):
    x = str(x)
    
    if z == 1:
        if len(x) > y:
            return 'long'
        else:
            return 'short'
    else:
        if len(x) < y:
            return 'long'
        else:
            return 'short'
```
